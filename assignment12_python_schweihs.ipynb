```python
import json
import pandas as pd
import numpy as np
workdir = "/Users/maggie/Documents/Code/DS710/ds710fall2016assignment12"

#MagicCardSet class stores all information about a set of Magic Cards 
class MagicCardSet(object):
    def __init__(self, name = "", magicCardsInfoCode = "", releaseDate = "",
                border = "", block = "", booster = "", translations = "", 
                 mkm_name = "", mkm_id = "", cards = []):
        self.name = name #object['name']
        #self.code = code #object["code"]
        self.InfoCode = magicCardsInfoCode #object["magicCardsInfoCode']
        self.releaseDate = releaseDate #object["releaseDate"]
        self.border = border #object["border"]
        self.type = type #object["type"]
        self.block = block #object["block"]
        self.booster = booster #object["booster"]
        self.translations = translations #object["translations"]
        self.mkm_name = mkm_name #object["mkm_name"]
        self.mkm_id = mkm_id #object["mkm_id"]
        self.cards = cards
    #Function to return information about the set of cards. Only one function was needed for the questions in
    #this problem, so only one function was defined.
    def get_cards(self):
        return self.cards
    
#Create an instance of the class which contains all of the information about the Dragons of Tarkir set   


with open("%s/DTK.json" % workdir) as data_file:    
    dtk = json.load(data_file)
    dtk_deck = MagicCardSet(dtk["name"], dtk["magicCardsInfoCode"], 
                 dtk["releaseDate"], dtk["border"], dtk["block"], dtk["booster"], 
                 dtk["translations"], dtk["mkm_name"], dtk["mkm_id"], dtk["cards"])
    dtk_card_list = dtk_deck.get_cards() #create a list of cards in JSON format
    dtk_cards = [] #Turn the list of JSON into a new list of MagicCard objects
    for card in dtk_card_list: #Not all cards contain all fields, if else statements check if the field exists
        dtk_cards.append(MagicCard(card["name"] if "name" in card else "", card["manaCost"] if "manaCost" in card else "",
                                  card["cmc"] if "cmc" in card else "", card["colors"] if "colors" in card else "",  
                                  card["rarity"] if "rarity" in card else "", card["subtypes"] if "subtypes" in card else "",
                                  card["supertypes"] if "supertypes" in card else "", card["text"] if "text" in card else "",
                                  card["flavor"] if "flavor" in card else "", card["artist"] if "artist" in card else "", 
                                  card["number"] if "number" in card else "", card["power"] if "power" in card else "", 
                                  card["toughness"] if "toughness" in card else "", card["layout"] if "layout" in card else "", 
                                  card["multiverseid"] if "multiversid" in card else "", card["imageName"] if "imageName" in card else "",
                                  card["id"] if "id" in card else ""))
```
